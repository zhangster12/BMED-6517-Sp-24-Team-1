```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pickle
import sys

parent_dir = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
normalized_data = parent_dir + "/data/interim/normalized"
sys.path.append(parent_dir)
```

```python
norm_data_paths = os.listdir(normalized_data)
norm_data_paths = [normalized_data + '/' + s for s in norm_data_paths]

for subject in norm_data_paths:
    # Create folder for each subject to save windowed data into
    save_dir = parent_dir + "/data/interim/10s_windows/"
    
    # Determine number of activities in subject normalized data
    # Read files in subject folder of normalized data
    activities = os.listdir(subject)
    
    # Open pkl file of activity
    for activity in activities:
        with open(subject + '/' + activity, 'rb') as f:
            data = pickle.load(f)
        
        # If not exist, create folders for subjects
        if not os.path.exists(save_dir + f"/{os.path.basename(os.path.normpath(subject))}/{activity[:-4]}"):
            os.makedirs(save_dir + f"/{os.path.basename(os.path.normpath(subject))}/{activity[:-4]}")
        
        idx = 0
        count = 0
        
        # 10 second window length (500Hz * 10)
        window_length = 5000
        
        # Sliding window length 
        slide = 500
        
        while (idx + window_length) < len(data['time']):
            end_idx = idx + window_length
            
            window={
                "time":None,
                "ecg":None,
                "ppg_ir_1":None,
                "ppg_r_1":None,
                "ppg_g_1":None,
                "ppg_ir_2":None,
                "ppg_r_2":None,
                "ppg_g_2":None,
                "scg_LT":None,
                "scg_HF":None,
                "scg_DV": None,
                "pcg_DV": None,
                "subject_id":None,
                "activity": None,
                "interval": None,
            }
            
            window['subject_id'] = os.path.basename(os.path.normpath(subject))
            window['activity'] = activity
            window['interval'] = [idx,end_idx]
            
            window['time'] = data['time'][idx:end_idx]
            window['ecg'] = data['ecg'][idx:end_idx]
            window['ppg_ir_1'] = data['ppg_ir_1'][idx:end_idx]
            window['ppg_r_1'] = data['ppg_r_1'][idx:end_idx]
            window['ppg_g_1'] = data['ppg_g_1'][idx:end_idx]
            window['ppg_ir_2'] = data['ppg_ir_2'][idx:end_idx]
            window['ppg_r_2'] = data['ppg_r_2'][idx:end_idx]
            window['ppg_g_2'] = data['ppg_g_2'][idx:end_idx]
            window['scg_LT'] = data['scg_LT'][idx:end_idx]
            window['scg_HF'] = data['scg_HF'][idx:end_idx]
            window['scg_DV'] = data['scg_DV'][idx:end_idx]
            window['pcg_DV'] = data['pcg_DV'][idx:end_idx]

            # Save pkl file of activity
            with open (f"{save_dir}/{os.path.basename(os.path.normpath(subject))}/{activity[:-4]}/{count}.pkl", "wb") as f:
                pickle.dump(window, f)
            
            idx = idx + slide
            count = count + 1
```
